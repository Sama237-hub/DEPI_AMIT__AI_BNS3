```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://wuzzuf.net/search/jobs/?q=data%20scientist&a=hpb'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
job_cards = soup.find_all('div', {'class': 'css-1gatmva'})
```

```python
data = []

for job in job_cards:
    try:
        company = job.find('a', {'class': 'css-17s97q8'}).text
        location = job.find('span', {'class': 'css-5wys0k'}).text
        date_posted = job.find('div', {'class': 'css-4c4ojb'}).text
        title = job.find('h2').text
        data.append([title, company, location, date_posted])
    except:
        continue
```

```python
df = pd.DataFrame(data, columns=['Title', 'Company', 'Location', 'Posted'])
df.head()
```

```python
df.to_csv('jobs.csv', index=False)
```
